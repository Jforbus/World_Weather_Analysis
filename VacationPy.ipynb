```python
# dependencies
import pandas as pd
import gmaps
import requests
#API key
from config import g_key
```

```python
# import dataframe
city_data_df = pd.read_csv('weather_data/cities.csv')
city_data_df.head()
```

```python
# configure gmaps to use api key
gmaps.configure(api_key=g_key)
```

```python
# remove negative temps

max_temp = city_data_df['Max Temp']

#temps = [max(temp,0) for temp in max_temp]

#for temp in max_temp:
    #temps.append(max(temp,0))

# create heatmap
# assign locations to an array of lat and lon pairs
locations = city_data_df[['Lat', 'Lng']]
# assign variable weights to array of equal length
max_temp = city_data_df['Max Temp']
# assign fig variable
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.7)
# assign heatmap_layer variable and pass in locations and weights, use list comprehension to remove negatives
heatmap_layer = gmaps.heatmap_layer(locations, weights=[max(temp,0) for temp in max_temp], dissipating=False, max_intensity=300, point_radius=4)
# add heatmap layer to figure
fig.add_layer(heatmap_layer)
# call figure
fig
```

```python
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
preferred_cities_df = city_data_df.loc[(city_data_df['Max Temp']<= max_temp)&(city_data_df['Max Temp']>= min_temp)]
preferred_cities_df.head(10)
```

```python
preferred_cities_df.count()
```

```python
hotel_df = preferred_cities_df[['City', 'Country', 'Max Temp', 'Lat', 'Lng']].copy()
hotel_df['Hotel Name'] = ''
hotel_df.head(10)
```

```python
# set parameters to search for a hotel
params = {
    'radius': 5000,
    'type' : 'lodging',
    'key' : g_key
}

# iterate through dataframe
for index, row in hotel_df.iterrows():
    
    # get the lat and lon
    lat = row['Lat']
    lng = row['Lng']
    
    # add coordinates to params
    params['location'] = f'{lat},{lng}'
    
    # create search URL
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    hotels = requests.get(base_url, params=params).json()
    
    # get first hotel name from request data
    try:
        
        hotel_df.loc[index, 'Hotel Name'] = hotels['results'][0]['name']
        
    except(IndexError):
        
        print('Hotel not found...skipping.')
```

```python
hotel_df.head(10)
```

```python
# Add a heatmap of temperature for the vacation spots.

# heat_layer variables
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]

# create figure
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

# create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp, dissipating=False,
             max_intensity=300, point_radius=4)

# create marker layer
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]

marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

# add layers to figure
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)
# Call the figure to plot the data.
fig
```
